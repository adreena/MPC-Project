```python
import matplotlib.pyplot as plt
```

```python
def plot(x,y, name):  
    fig = plt.figure()
    plt.plot(x, y, 'r')
    plt.show()
    fig.savefig('{}.png'.format(name))
```

```python
cte_vals = []

print('CTE')
with open("./cte_vals.txt", 'r') as cte:
    for line in cte:
        cte_vals.append(float(line.split('\n')[0]))
x_data = list(range(0,len(cte_vals)))
plot(x_data,cte_vals,'cte')
print('-------------------------------')


# print('V')
# v_vals = []
# with open("./v_vals.txt", 'r') as cte:
#     for line in cte:
#         v_vals.append(float(line.split('\n')[0]))
# x_data = list(range(0,len(v_vals)))
# plot(x_data,v_vals,'v')
# print('-------------------------------')


print('Angle')
delta_vals = []
with open("./delta_vals.txt", 'r') as cte:
    for line in cte:
        delta_vals.append(float(line.split('\n')[0]))
x_data = list(range(0,len(delta_vals)))
plot(x_data,delta_vals,'delta')
print('-------------------------------')
```
